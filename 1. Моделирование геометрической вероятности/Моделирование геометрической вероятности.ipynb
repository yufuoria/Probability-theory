<center><h1>Моделирование геометрической вероятности</h1></center>

<b><center><h2>Задача</h2></center></b>

Имеется прямоугольник шириной 4 единицы и высотой 2 единицы. Внутри этого прямоугольника находится треугольник с вершинами в точках (0, 0), (4, 1) и (2, 0). Необходимо определить вероятность того, что случайно выбранная точка внутри прямоугольника окажется внутри треугольника.

<b><center><h3>Теоретический расчёт</h3></center></b>

Рассчитаем площадь прямоугольника: $$S_{прям}=ab=4 \cdot 2=8$$

Рассчитаем площадь треугольника: $$S_{треуг}=\frac{1}{2} |(x_{2}-x_{1})(y_{3}-y_{1})-(x_{3}-x_{1})(y_{2}-y_{1})|=\frac{1}{2}|(4-0)(0-0)-(2-0)(1-0)|=1$$

Тогда вероятность того, что случайно выбранная точка внутри прямоугольника окажется внутри треугольника: $$P(A)=\frac{S_{треуг}}{S_{прямоуг}}=\frac{1}{8}=0.125$$

<b><center><h3>Моделирование</h3></center></b>

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon
import numpy as np
```

```python
def points(number):
    a = 4
    b = 2
    
    fig, ax = plt.subplots(figsize=(5, 5))
    ax = plt.gca()
    ax.add_patch(Rectangle((0, 0), a, b, facecolor='purple'))
    ax.add_patch(Polygon([[0, 0], [2, 0], [4, 1]], facecolor='pink'))
    
    x = np.random.uniform(0, a, number)
    y = np.random.uniform(0, b, number)
    plt.scatter(x, y, marker='.', c='black')
    
    ls = []
    for i in range(len(x)):
        if (y[i] >= 0 and y[i] <= 0.25 * x[i] and y[i] >= 0.5 * x[i] - 1):
            ls.append((x[i], y[i]))
    inside = len(ls)
    probability = inside / number
    
    plt.show()
    print(f'Внутрь треугольника попало {inside} точек из {number}. Доля попавших: {probability}')    
```

```python
points(1000)
```
